```python
import pandas as pd
import numpy as np

# Random sayılarla 5'e 3'lük bir numpy arrayi elde ettik.
# Bu değerleri kullanarak birinci data frame'i oluşturduk.
# Bu değerlere 99 ekleyerek ikinci data frame'i oluşturduk.

m = np.random.randint(1,30, size=(5,3))
df1 = pd.DataFrame(m, columns=["var1","var2","var3"])
df2 = df1 + 99
```

```python
# İki dataframe'i alt alta birleştirdik
pd.concat([df1,df2]) 
```

```python
# İndexler baştan başlar
pd.concat([df1,df2], ignore_index=True) 
```

```python
# Merge ile birleştirme
df1= pd.DataFrame({"employees":["john","dennis","mark","maria"],"group":["accounting","engineering","engineering","hr"]})

df2= pd.DataFrame({"employees":["mark","john","dennis","maria"],"start_date":[2010,2009,2014,2019]})
```

```python
# Neye göre birleşecek belirtebiliriz.
# pd.merge(df1,df2, on="employees") 

# İki df'yi birleştirdi bunu çalışana göre yaptı
df3 = pd.merge(df1,df2) 
```

```python
df4 = pd.DataFrame({"group": ["accounting","engineering","hr"],"manager":["Caner","Mustafa","Berkcan"]})
```

```python
# İki df'yi birleştirdi bunu grup bilgisine göre yaptı
pd.merge(df3,df4) 
```
