```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Plot özelliği -> Veri Görselleştirme
x = np.array([1,8])
y = np.array([0,150])

plt.plot(x,y)
plt.show() # Çizgi çekti
```

```python
plt.plot(x,y, 'o')
plt.show() # Nokta koydu
```

```python
x = np.array([2,4,6,8,10])
y = np.array([1,3,5,7,9])

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'o')
plt.show()
```

```python
# Marker
y = np.array([13,28,11,100])

plt.plot(y,marker='o')
plt.show()
```

```python
# Line 
y = np.array([13,28,11,100])

plt.plot(y, linestyle="dashed") #dotted,dashdot
plt.show()
```

```python
plt.plot(y, linestyle="dashed",color="r")
plt.show()
```

```python
# Multiple Lines
x = np.array([23,18,31,10])
y = np.array([13,28,11,100])

plt.plot(x)
plt.plot(y)
plt.show() # show demeden de çalışıyor hepsi ama alışkanlık
```

```python
# Labels
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(x,y)
# Başlık
plt.title("Ana Başlık")
# X ekseni isimlendirme
plt.xlabel("X ekseni")
# Y ekseni isimlendirme
plt.ylabel("Y ekseni")
# Izgara
plt.grid()
```

```python
# Subplots -> birden fazla grafik birleştirme

# plot 1
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])

plt.subplot(1,2,1) # 1 Satırlık 2 Sutünlük 1.sini verdim
plt.title("1")
plt.plot(x,y)

# plot 2
x = np.array([8,8,9,9,10,15,11,15,12,15])
y = np.array([24,20,26,27,280,29,30,30,30,30])

plt.subplot(1,2,2) # 1 Satırlık 2 Sutünlük 2.sini verdim
plt.title("2")
plt.plot(x,y)
```
