```python
import pandas as pd
import seaborn as sns
df = sns.load_dataset("titanic")
```

```python
df.pivot_table("survived","sex","embarked") 
# Cinsiyet ve gemiye binme kesişiminde hayatta kalma ortalaması
# Ön tanımlı fonksiyonu ortalamadır.
```

```python
df.pivot_table("survived","sex","embarked", aggfunc="std") 
# Fonksiyonu standart sapma olarak değiştirdik
```

```python
df.pivot_table("survived","sex",["embarked","class"]) 
# Class değişkeni, gemiye binme limanı ile birlikte kullandık
```

```python
df["new_age"] = pd.cut(df["age"],[0,10,18,25,40,90])
# Yaş değişkenini kategorize etmek için cut veya qcut kullanılır. Sonda verilen listeye göre kategorize edilir
```

```python
df.pivot_table("survived","sex","new_age")
# Yaş kategorisine göre hayatta kalma değerlerini inceledik.
```

```python
df.pivot_table("survived","sex",["new_age","class"])
# Hem sınıf hem de yaşa göre sorguladık.
```
