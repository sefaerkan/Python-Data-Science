```python
import pandas as pd
import seaborn as sns
df = sns.load_dataset("titanic")
```

```python
df["age"].mean() # Yaş ortalaması
```

```python
df.groupby("sex")["age"].mean() # Kadın erkek yaş ortalaması -> cinsiyete göre grupla, yaş ortalamasını al
```

```python
df.groupby("sex").agg({"age":"mean"}) # Kadın erkek yaş ortalaması
```

```python
df.groupby("sex").agg({"age":["mean","sum"]}) # Kadın erkek yaş ortalaması ve toplamı
```

```python
df.groupby("sex").agg({"age":["mean","sum"],"survived":"mean"}) # Hayatta kalma değerlerinin ortalamasını aldık.
```

```python
df.groupby(["sex","embark_town"]).agg({"age":"mean","survived":"mean"}) 
# Cinsiyet ve bindiği limana göre sınıflandırıp yaş ortalaması ve hayatta kalma değeri ortlaması aldık
```

```python
df.groupby(["sex","embark_town","class"]).agg({"age":"mean","survived":"mean"}) 
# Üstteki örneğe ek olarak class ekledik
```

```python
df.groupby(["sex","embark_town","class"]).agg({"age":"mean","survived":"mean","sex":"count"}) # Ek olarak yaş sayısını ekledik
```
