```python
# Görev 16: 30 yaşın altında olanlar 1, 30'a eşit ve üstünde olanlara 0 verecek bir fonksiyon yazın. Yazdığınız fonksiyonu kullanarak titanik veri setinde age_flag adında bir değişken oluşturunuz.
def age_flag_func(age):
    if age < 30:
        return 1
    else:
        return 0

def c(dataframe, col):
    dataframe['age_flag'] = dataframe[col].apply(age_flag_func)
    return dataframe
                          
df = c(df, 'age')
print(df)                     
```

```python
# Görev 17: Seaborn kütüphanesi içerisinden Tips veri setini tanımlayınız.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset("tips")
```

```python
# Görev 18: Time değişkeninin kategorilerine (Dinner, Lunch) göre total_bill değerinin sum, min, max ve mean değerleri
df.groupby("time").agg({"total_bill":["sum","min","max","mean"]})
```

```python
# Görev 19: Day ve time’a göre total_bill değerlerinin sum, min, max ve mean değerleri
df.groupby(["day","time"]).agg({"total_bill":["sum","min","max","mean"]})
```

```python
# Görev 20: Lunch zamanına ve kadın müşterilere ait total_bill ve tip değerlerinin day'e göre sum, min, max ve mean değerleri
filtered_df = df[(df['time'] == 'Lunch') & (df['sex'] == 'Female')]

filtered_df.groupby('day')[['total_bill', 'tip']].agg(['sum', 'min', 'max', 'mean'])
```

```python
# Görev 21: size'i 3'ten küçük, total_bill'i 10'dan büyük olan siparişlerin ortalaması nedir? (loc kullanınız)
df.loc[(df["size"] < 3) & (df["total_bill"] > 10)]
```

```python
# Görev 22: total_bill_tip_sum adında yeni bir değişken oluşturunuz. Her bir müşterinin ödediği totalbill ve tip in toplamını versin
df["total_bill_tip_sum"] = df["total_bill"] + df["tip"]
```

```python
"""" Görev 23: Total_bill değişkeninin kadın ve erkek için ayrı ayrı ortalamasını bulunuz.
Bulduğunuz ortalamaların altında olanlara 0, üstünde ve eşit
olanlara 1 verildiği yeni bir total_bill_flag değişkeni oluşturunuz.
Kadınlar için Female olanlarının ortalamaları, erkekler için ise Male olanların ortalamaları dikkate alınacktır.
Parametre olarak cinsiyet ve total_bill
alan bir fonksiyon yazarak başlayınız. (If-else koşulları içerecek) """

def create_total_bill_flag(df, gender_col, total_bill_col):
    # Cinsiyete göre gruplayarak total_bill'in ortalamasını bulma
    gender_mean = df.groupby(gender_col)[total_bill_col].mean()
    
    # Kadın ve erkek ortalamalarını al
    female_mean = gender_mean['Female']
    male_mean = gender_mean['Male']
    
    # total_bill_flag oluşturma
    df['total_bill_flag'] = (df[total_bill_col] >= female_mean).astype(int) if df[gender_col].iloc[0] == 'Female' else (df[total_bill_col] >= male_mean).astype(int)
    
    return df


df = create_total_bill_flag(df, 'sex', 'total_bill')

print(df)
```

```python
# Görev 24: total_bill_flag değişkenini kullanarak cinsiyetlere göre ortalamanın altında ve üstünde olanların sayısını gözlemleyiniz.
df.groupby('sex')['total_bill_flag'].value_counts()
```

```python
# Görev 25: Veriyi total_bill_tip_sum değişkenine göre büyükten küçüğe sıralayınız ve ilk 30 kişiyi yeni bir dataframe'e atayınız

# total_bill_tip_sum değişkenine göre büyükten küçüğe sıralama yapın ve ilk 30 kişiyi seçin
top_30 = df.sort_values(by='total_bill_tip_sum', ascending=False).head(30).copy()

# Yeni bir veri çerçevesine atayın
df_top_30 = top_30.copy()

print(df_top_30)
```
