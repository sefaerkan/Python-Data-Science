```python
# Kategorik değişkenler için: Sutün grafik, countplot bar

# Sayısal değişkenler için: hist, boxplot
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset("titanic")
```

```python
df["sex"].value_counts().plot(kind="bar") # Sutün Grafik - Bar
plt.show()
```

```python
plt.hist(df["age"]) # Sayısal Veri -> Histogram Sutün
plt.show()
```

```python
plt.boxplot(df["fare"]) # Sayısal Veri -> Kutu Grafik 
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
cb = sns.load_dataset("tips")
```

```python
cb["sex"].value_counts()
```

```python
sns.countplot(x=cb["sex"],data=cb) # Seaborn kategorik
plt.show()
```

```python
sns.boxplot(x=cb["total_bill"]) # Kutu Grafik
plt.show()
```

```python
cb["total_bill"].hist() # Histogram
plt.show()
```
